(a) Three-dimensional plots:

```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection = '3d')

# Function f1(x,y) = x**2+4*y**2-2*x**2*y+4 with minimum p1 = (0,0,4)
def f(x, y):
  return (x**2+4*y**2-2*x**2*y+4)

x = np.linspace(-1, 1, 1000)
y = np.linspace(-1, 1, 1000)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.set_title('Function f1 with minimum p1')
ax.contour3D(Y, X, Z, 50, cmap='binary')
ax.set_xlabel('y')
ax.set_ylabel('x')
ax.set_zlabel('z')

ax.plot([0], [0], [4], markerfacecolor='Red', markeredgecolor='k', marker='o', markersize=10, alpha=1)

plt.show()
```

```
fig = plt.figure()
ax = plt.axes(projection = '3d')

# Function f2(x,y) = 3*x**2*y+y**3-3*x**2-3*y**2+2 with maximum p1 = (0,0,2)
def f(x, y):
  return (3*x**2*y+y**3-3*x**2-3*y**2+2)

x = np.linspace(-1, 1, 1000)
y = np.linspace(-1, 1, 1000)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.set_title('Function f2 with maximum p2')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.plot([0], [0], [2], markerfacecolor='Red', markeredgecolor='k', marker='o', markersize=10, alpha=1)

plt.show()
```

```
fig = plt.figure()
ax = plt.axes(projection = '3d')

# Function f3(x,y) = 3*x**2*y+y**3-3*x**2-3*y**2+2 with saddle p3 = (1,1,0)
def f(x, y):
  return (3*x**2*y+y**3-3*x**2-3*y**2+2)

x = np.linspace(0, 1, 1000)
y = np.linspace(0, 2, 1000)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.set_title('Function f3 with saddle point p3')
ax.contour3D(Y, X, Z, 50, cmap='binary')
ax.set_xlabel('y')
ax.set_ylabel('x')
ax.set_zlabel('z')

ax.plot([1], [1], [0], markerfacecolor='Red', markeredgecolor='k', marker='o', markersize=10, alpha=1)

plt.show()
```
